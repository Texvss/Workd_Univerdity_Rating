```python
plt.figure(figsize=(16, 10))

plt.subplot(321)  # размер и положение подграфика
plt.hist(df.Age)
plt.title("Распределение по возрасту")
plt.grid(True)  # добавить сетку

plt.subplot(322)
plt.bar(df.Sex.value_counts().index, df.Sex.value_counts().values)
plt.title("Количество м / ж")

plt.subplot(323)
plt.hist(df.Age, color="yellow")
plt.title("Распределение по возрасту")

plt.subplot(324)
plt.bar(df.Sex.value_counts().index, df.Sex.value_counts().values, color="cyan")
plt.title("Количество м / ж")

plt.subplot(313)
scatter = plt.scatter(df.Fare, df.Age, c=df.Survived, s=df.Pclass)
plt.title("Выживаемость относительно цены билета / возраста и класса")


plt.savefig("stack.png")  # сохранить изображение
```

```python
# https://matplotlib.org/3.1.1/gallery/subplots_axes_and_figures/figure_title.html


def f(t):
    s1 = np.cos(2 * np.pi * t)
    e1 = np.exp(-t)
    return s1 * e1


t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
t3 = np.arange(0.0, 2.0, 0.01)


fig, axs = plt.subplots(2, 1, figsize=(16, 10))

axs[0].plot(t1, f(t1), "o", t2, f(t2), "-")
axs[0].set_title("subplot 1")
axs[0].set_xlabel("distance (m)")
axs[0].set_ylabel("Damped oscillation")
fig.suptitle("This is a somewhat long figure title", fontsize=16)

axs[1].plot(t3, np.cos(2 * np.pi * t3), "--")
axs[1].set_xlabel("time (s)")
axs[1].set_title("subplot 2")
axs[1].set_ylabel("Undamped")

plt.show()
```

# Plotly
https://plotly.com/python/

```python
import plotly.graph_objects as go

_tdata = df.Pclass.value_counts()

labels = _tdata.index
values = _tdata.values

go.Figure(data=[go.Pie(labels=labels, values=values)])
```

```python
import plotly.express as px

px.pie(df, values="Survived", names="Pclass", title="Fare / Pclass", hole=0.3)
```

```python
df["SurvivedW"] = df.Survived.apply(lambda x: "Выжил" if x else "Погиб")
```

```python
px.scatter(
    df[:50],
    x="Fare",
    y="Age",
    color="SurvivedW",
    size="Pclass",
    symbol="Embarked",
    hover_data=["Sex", "Name"],
)
```

```python
px.line(
    df,
    x="PassengerId",
    y="Fare",
    color="SurvivedW",
)
```

```python
px.sunburst(df.dropna(), path=["Pclass", "Parch", "Sex"], values="Survived")
```

```python
px.bar(
    x=df["Pclass"].value_counts().index,
    y=df["Pclass"].value_counts().values,
)
```
